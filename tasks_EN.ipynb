# Task 5: Calculate the number of orders for each user (`CustomerID`) from Germany. Keep only those who made more than N transactions (`InvoiceNo`), where N is the 80th percentile. Save the resulting user IDs in `germany_top` (just the IDs, not the entire DataFrame).

<font size="5">The filtered data is stored in `retail`. The order identifier is `InvoiceNo`. Each order can have more than one line in the data.</font>


```python
# Filter data by country "Germany" to keep only orders from Germany.
# Group data by 'CustomerID' to count the unique orders for each user.
# Calculate the count of unique orders ('InvoiceNo') for each user.
# Rename the 'InvoiceNo' column to 'Orders_count' for readability.
# Keep only those users who made more orders than the 80th percentile in order count.  
# Select only the user IDs (`CustomerID`) from the filtered data..
germany_top = retail_no_duplicates.query('Country == "Germany"') \
    .groupby('CustomerID', as_index=False) \
    .agg({'InvoiceNo': 'nunique'}) \
    .rename(columns={'InvoiceNo': 'Orders_count'}) \
    .query('Orders_count > Orders_count.quantile(0.8)').CustomerID
```

# Task 6: Using the object with user IDs (`germany_top`) from the previous step, filter the data to retain records only for the relevant users. Save the resulting DataFrame as `top_retail_germany`.

<font size="5">DataFrame is saved in `retail`, the necessary user IDs are in `germany_top`.</font>


# Answer:

```python
top_retail_germany = retail.query('CustomerID in @germany_top')
top_retail_germany
```

# Task 7: Group `top_retail_germany` by product code (`StockCode`). Which product was added to the cart most frequently, excluding POST?

```python
# Filter data, excluding the product with code "POST"
# Group the filtered data by product code ('StockCode') 
# Count the number of transactions for each product (by the number of unique 'InvoiceNo')  
# Sort products by transaction count in descending order
top_retail_germany.query('StockCode != "POST"') \
    .groupby('StockCode', as_index=False) \
    .agg({'InvoiceNo': 'count'}) \
    .sort_values('InvoiceNo', ascending =False)
```

# Task 8: Return to analyzing the complete dataset `retail`. Create a `Revenue` column with the purchase amount using `Quantity` and `UnitPrice`.

```python
retail['Revenue'] = retail.Quantity * retail.UnitPrice
retail
```

# Task 9: For each transaction (`InvoiceNo`), calculate the final order amount. Specify the top 5 transactions (`InvoiceNo`) by order amount as the answer.

```python
# Group data by transaction identifier ('InvoiceNo') 
# Sum values in the 'Revenue' column for each transaction (final order amount)  
# Sort transactions by final order amount ('Revenue') in descending order
# Select the top 5 transactions with the highest order amounts 
retail.groupby('InvoiceNo', as_index=False) \
    .agg({'Revenue': 'sum'}) \
    .sort_values('Revenue', ascending=False).head(5)
```
