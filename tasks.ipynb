# Задача 5 "Посчитайте число заказов для каждого пользователя (CustomerID) из Германии (Germany). Оставьте только тех, кто совершил более N транзакций (InvoiceNo), где N – 80% процентиль. Запишите полученные id пользователей в germany_top (не весь датафрейм, только id)."

<font size="5">Отфильтрованные данные записаны в retail. Идентификатор заказа – InvoiceNo. Для каждого заказа в данных может встречаться более 1 строки</font>

```python
# Фильтруем данные по стране "Germany" и оставляем только заказы из Германии.
# Группируем данные по 'CustomerID', чтобы посчитать количество уникальных заказов для каждого пользователя.
# Считаем количество уникальных заказов ('InvoiceNo') для каждого пользователя.
# Переименовываем колонку 'InvoiceNo' в 'Orders_count' для лучшей читаемости.
# Оставляем только тех пользователей, которые совершили больше заказов, чем 80-й процентиль по количеству заказов.
# Из полученных данных выбираем только идентификаторы пользователей (CustomerID).
germany_top = retail_no_duplicates.query('Country == "Germany"') \
    .groupby('CustomerID', as_index=False) \
    .agg({'InvoiceNo': 'nunique'}) \
    .rename(columns={'InvoiceNo': 'Orders_count'}) \
    .query('Orders_count > Orders_count.quantile(0.8)').CustomerID
```

# Ответ

```python
germany_top
```

# Задача 6 "Используя объект с id пользователей (germany_top), полученный на предыдущем шаге, отфильтруйте наблюдения и оставьте в данных записи только по интересующим нас юзерам. Результирующий датафрейм запишите в top_retail_germany."

<font size="5">Датафрейм записан в retail, нужные id пользователей – в germany_top</font>

# Ответ

```python
top_retail_germany = retail.query('CustomerID in @germany_top')
top_retail_germany
```

# Задача 7 "Сгруппируйте top_retail_germany по коду товара (StockCode). Какой из продуктов добавляли в корзину чаще всего, кроме POST?"

```python
# Отфильтровываем данные, исключая продукт с кодом "POST"
# Группируем отфильтрованные данные по коду товара ('StockCode')
# Считаем количество транзакций для каждого товара (по числу уникальных 'InvoiceNo')
# Сортируем товары по количеству транзакций в порядке убывания
top_retail_germany.query('StockCode != "POST"') \
    .groupby('StockCode', as_index=False) \
    .agg({'InvoiceNo': 'count'}) \
    .sort_values('InvoiceNo', ascending =False)
```

# Задача 8 "Вернемся к анализу полного датасета retail. Создайте колонку Revenue с суммой покупки, используя колонки Quantity и UnitPrice."

# Задача 9 "Для каждой транзакции (InvoiceNo), посчитайте финальную сумму заказа. В качестве ответа укажите топ-5 (InvoiceNo) по сумме заказа"

```python
# Группируем данные по идентификатору транзакции ('InvoiceNo')
# Для каждой транзакции суммируем значения по колонке 'Revenue' (финальная сумма заказа)
# Сортируем транзакции по финальной сумме заказа ('Revenue') в порядке убывания
# Отбираем топ-5 транзакций с наибольшей суммой заказа
retail.groupby('InvoiceNo', as_index=False) \
    .agg({'Revenue': 'sum'}) \
    .sort_values('Revenue', ascending=False).head(5)
```
